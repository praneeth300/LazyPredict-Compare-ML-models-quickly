__Lazypredict__ is a Open source python package created by shankar rao pandala.Developement and contribution to this are still going. As, of Now it only based on Supervised techniques known as Regression and classification.

While building machine learning models we are not sure which algorithm should work well with the given dataset, hence we end up trying many models and keep iterating until we get proper accuracy. Have you ever thought about getting all the basic algorithms at once to predict for model performance? 

LazyPredict is a module helpful for this purpose. LazyPredict will generate all the basic machine learning algorithmsâ€™ performances on your model. Along with the accuracy score, LazyPredict provides certain evaluation metrics and the time taken by each model.

__Classification__

```
#Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import lazypredict
from lazypredict.Supervised import LazyClassifier
```

```
#Get the data from sklearn
from sklearn import datasets
df=datasets.load_breast_cancer()
```

```
df_1=pd.DataFrame(df.data)
df_1.columns=df.feature_names
df_1['Target']=df.target
df_1.head()
```

```
df_1['Target'].value_counts()
```

Check the data is balanced or not

```
df_1.isnull().sum()
```

There is no missing values are present in the data,Then we go 

```
#Split the data in to x and y
x=df_1.drop('Target',1)
y=df_1['Target']

#Splitting the dataset in to train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)

#Build the model
lazy=LazyClassifier(verbose=0,ignore_warnings=True,custom_metric=None)

train,test=lazy.fit(x_train,x_test,y_train,y_test)

#Get the model performance

train
```

```
test
```

```
#Visualize the performance by using barplot
plt.figure(figsize=(15,7))
ax=sns.barplot(x=train.index,y=train['Accuracy'])
ax.set_title('Model Performance',fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(8,10))
ax=sns.barplot(y=test.index,x=test['Accuracy'])
ax.set_title('Model Performance',fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

__Regression__

```
#Get the data
reg=datasets.load_boston()
reg_1=pd.DataFrame(reg.data)
reg_1.columns=reg.feature_names
reg_1['Target']=reg.target
```

```
reg_1.head()
```

```
#Import the libarary
from lazypredict.Supervised import LazyRegressor
#initalize a model
lazy_reg=LazyRegressor(verbose=0,custom_metric=None,ignore_warnings=True)

x=reg_1.drop('Target',1)
y=reg_1['Target']

#Split the data in to train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#Fit the model
train,test=lazy_reg.fit(x_train,x_test,y_train,y_test)

#Get the performance
train
```

```
test
```

```
plt.figure(figsize=(10, 5))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x=train.index, y="R-Squared", data=train)
ax.set(ylim=(0, 1))
plt.xticks(rotation=90)
```

```
plt.figure(figsize=(10, 12))
sns.set_theme(style="whitegrid")
ax = sns.barplot(y=test.index, x="R-Squared", data=test)
ax.set(xlim=(0, 1))
plt.xticks(rotation=90)
```
